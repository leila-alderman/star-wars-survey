# Star Wars Survey

While waiting for *Star Wars: The Force Awakens* to come out, the team at FiveThirtyEight became interested in answering some questions about Star Wars fans. In particular, they wondered: does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?

The team needed to collect data addressing this question. To do this, they surveyed Star Wars fans using the online tool SurveyMonkey, and they received 835 total responses.

The data has several columns, including:

- `RespondentID` - An anonymized ID for the respondent (person taking the survey)
- `Gender` - The respondent's gender
- `Age` - The respondent's age
- `Household Income` - The respondent's income
- `Education` - The respondent's education level
- `Location` (Census Region) - The respondent's location
- `Have you seen any of the 6 films in the Star Wars franchise?` - Has a `Yes` or `No` response
- `Do you consider yourself to be a fan of the Star Wars film franchise?` - Has a `Yes` or `No` response

There are several other columns containing answers to questions about the Star Wars movies. For some questions, the respondent had to check one or more boxes. 

```python
# import pandas
import pandas as pd
# import the data with the correct encoding
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
# Start by exploring the data.
star_wars.head(10)
```

```python
# Explore the column names.
star_wars.columns
```

```python
# The first row includes important information about the 
# questions and is needed to understand the columns.
print(star_wars.iloc[0])
```

```python
# Remove all rows where RespondentID is NaN.
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.head()
```

```python
# Convert the answers to the Yes/No questions to Boolean
yes_no = {
    'Yes' : True,
    'No' : False
}

# Rename the columns to be more manageable.
star_wars.rename(columns= {
    'Have you seen any of the 6 films in the Star Wars franchise?': 'seen_any',
    'Do you consider yourself to be a fan of the Star Wars film franchise?' : 'fan',
    'Which of the following Star Wars films have you seen? Please select all that apply.' : 'seen_ep1',
    'Unnamed: 4' : 'seen_ep2',
    'Unnamed: 5' : 'seen_ep3',
    'Unnamed: 6' : 'seen_ep4',
    'Unnamed: 7' : 'seen_ep5',
    'Unnamed: 8' : 'seen_ep6'
}, inplace=True)

star_wars['seen_any'] = star_wars['seen_any'].map(yes_no)
star_wars['fan'] = star_wars['fan'].map(yes_no)
star_wars.head()
```

```python
# Change the values in the 'seen_ep#' columns to either
# True or False.
import numpy as np

star_wars['seen_ep1'] = star_wars['seen_ep1'].map({
    'Star Wars: Episode I  The Phantom Menace': True,
    np.NaN: False
})

star_wars['seen_ep2'] = star_wars['seen_ep2'].map({
    'Star Wars: Episode II  Attack of the Clones': True,
    np.NaN: False
})

star_wars['seen_ep3'] = star_wars['seen_ep3'].map({
    'Star Wars: Episode III  Revenge of the Sith': True,
    np.NaN: False
})

star_wars['seen_ep4'] = star_wars['seen_ep4'].map({
    'Star Wars: Episode IV  A New Hope': True,
    np.NaN: False
})

star_wars['seen_ep5'] = star_wars['seen_ep5'].map({
    'Star Wars: Episode V The Empire Strikes Back': True,
    np.NaN: False
})

star_wars['seen_ep6'] = star_wars['seen_ep6'].map({
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
})

star_wars.head()
```

```python
# Convert the columns with the movie ratings to float type
star_wars[star_wars.columns[9:15]] = \
star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Rename more columns
star_wars.rename(columns= {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'rank_ep1',
    'Unnamed: 10': 'rank_ep2',
    'Unnamed: 11': 'rank_ep3',
    'Unnamed: 12': 'rank_ep4',
    'Unnamed: 13': 'rank_ep5',
    'Unnamed: 14': 'rank_ep6'
}, inplace=True)
star_wars.dtypes
```

```python
# Calculate the means of the ranking columns.
rank_cols = ['rank_ep1', 'rank_ep2', 'rank_ep3', 'rank_ep4', 'rank_ep5', 'rank_ep6']
ranks = star_wars[rank_cols].mean()

# Plot the mean ranks on a bar chart.
# Call the matplotlib magic.
%matplotlib inline
ranks.plot.bar()
```

The rankings show that Episode V is the overall favorite movie, while Episode III is the least favorite. As expected, the three original movies are all more highly ranked than the three later movies.

```python
# Compute the number of people who have seen each movie.
seen_cols = ['seen_ep1', 'seen_ep2', 'seen_ep3', 'seen_ep4', 'seen_ep5', 'seen_ep6']
seen_sums = star_wars[seen_cols].sum()

# Plot the sum of people who have seen each movie on a bar chart.
seen_sums.plot.bar()
```

The number of people who have seen the movies correlates reasonably well with the rankings: Episode V has been seen by the most people, and Episode III has been seen by the least. However, more people have seen Episode I than Episode IV, even though the rankings indicate that Episode IV was better.

```python
# Separate the data into fans and not fans to see if 
# there are any differences between the two groups.
fans = star_wars[star_wars['fan'] == True]
not_fans = star_wars[star_wars['fan'] == False]

ranks_fans = fans[rank_cols].mean()
ranks_not_fans = not_fans[rank_cols].mean()
df_rfan = pd.DataFrame({'fans': ranks_fans,
                        'not_fans': ranks_not_fans})
df_rfan.plot.bar()
```

```python
seen_fans = fans[seen_cols].sum()
seen_not_fans = not_fans[seen_cols].sum()
df_sfan = pd.DataFrame({'fans': seen_fans,
                        'not_fans': seen_not_fans})
df_sfan.plot.bar()
```

Unsurprisingly, the people who consider themselves fans of Star Wars are far more likely to have seen all of the movies than those who do not self-identify as Star Wars fans. Those who are not fans are most likely to have seen Episode V or VI, and approximately equal numbers of non-fans have watched Episodes II, III, and IV. 

Both fans and non-fans agree that Episode V is the best Star Wars movie. However, fans say that the second best movie is Episode VI, whereas non-fans say that Episode I is about as good as Episode V. Both groups agree that Episode III is the worst Star Wars movie.
